```python
from collections import deque
def add_node(v):
    if v in graph:
        print(v,"it is already present in node")
    else:
        graph[v]=[]
def add_edges(v1,v2):
    if v1 not in graph:
        print(v1,"it is not present in graph")
    if v2 not in graph:
        print(v2,"it is not present in graph")
    else:
        graph[v1].append(v2)
        graph[v2].append(v1)
def DFS(v,visited,graph):
    if v not in graph:
        print(v,"it is not present Graph")
    else:
        if v not in visited:
            print(v,end=" ")
            visited.add(v)
            for i in graph[v]:
                DFS(i,visited,graph)
def BFS(v,visited1,graph):
    if v not in graph:
        print(v,"it is not present in graph")
    else:
        queue=[]
        queue.append(v)
        visited1.add(v)
        while queue:
            current=queue.pop(0)
            print(current,end=" ")
            for i in graph[current]:
                if i not in visited1:
                    queue.append(i)
                    visited1.add(i)
def shortest_path(node,target,graph):
    if node not in graph:
        print(node,"it is not present in graph")
    elif target not in graph:
        print(target,"it is not present in graph")
    else:
        visited={}
        queue=deque()
        queue.append(node)
        visited[node]=None
        while queue:
            current=queue.popleft()
            if current==target:
                path=[]
                while current:
                    path.append(current)
                    current=visited[current]
                return path[::-1]
            for i in graph[node]:
                if i not in visited:
                    visited[i]=current
                    queue.append(i)
        
visited1=set()
visited=set()
graph={}
add_node("A")
add_node("B")
add_node("C")
add_node("E")
add_node("F")
add_node("G")
add_edges("A","B")
add_edges("B","C")
add_edges("A","C")
add_edges("C","G")
add_edges("G","F")
add_edges("A","E")
add_edges("B","F")
print(graph)
print('-------------********DFS*******-------------------')
DFS("A",visited,graph)
print()
print('------------------*******BFS******------------------')
BFS("A",visited1,graph)
print()
print("************Shortest Path************")
shortest_path("A","G",graph)
```
