```python
def add_node(v):
    global node_count
    if v in node:
        print(v,"it is already present in node")
    else:
        node_count=node_count+1
        node.append(v)
        for i in graph:
            i.append(0)
        temp=[]
        for i in range(node_count):
            temp.append(0)
        graph.append(temp)
def add_edge(v1,v2):
    if v1 not in node:
        print(v1,"it is not present in node")
    if v2 not in node:
        print(v2,"it is not present in node")
    else:
        index1=node.index(v1)
        index2=node.index(v2)
        graph[index1][index2]=1
        graph[index2][index1]=1
def delete_node(v):
    global node_count
    if v not in node:
        print(v,"it is not present in node")
    else:
        index1=node.index(v)
        node_count=node_count-1
        node.remove(v)
        graph.pop(index1)
        for i in graph:
            i.pop(index1)
def delete_edge(v1,v2):
    if v1 not in node:
        print(v1,"it is not in present in graph")
    if v2 not in node:
        print(v2,"it is not present in node")
    else:
        index1=node.index(v1)
        index2=node.index(v2)
        graph[index1][index2]=0
        graph[index2][index1]=0
def display():
    for i in range(node_count):
        for j in range(node_count):
            print(graph[i][j],end=" ")
        print()
node=[]
graph=[]
node_count=0
add_node("A")
add_node("B")
add_node("C")
add_node("D")
print(node)
print(graph)
display()
print('-------------------------')
add_edge("A","B")
add_edge("A","D")
add_edge("C","B")
add_edge("D","C")
add_edge("B","D")
display()
print('-------------------------')
delete_node("A")
display()
print('-------------------')
delete_edge("B","C")
display()
```
