```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
class CLL:
    def __init__(self):
        self.head=None
        self.tail=None
    def insert_front(self,data):
        nb=Node(data)
        if self.head is None:
            self.head=nb
            self.tail=nb
            self.tail.next=self.head
        else:
            nb.next=self.head
            self.head=nb
            self.tail.next=self.head
    def insert_end(self,data):
        ne=Node(data)
        if self.head is None:
            self.head=ne
            self.tail=ne
            self.tail.next=self.head
        else:
            self.tail.next=ne
            self.tail=ne
            self.tail.next=self.head
    def insert_position(self,pos,data):
        np=Node(data)
        if pos==1:
            self.insert_front(data)
        else:
            temp=self.head
            for i in range(1,pos-1):
                temp=temp.next
            np.next=temp.next
            temp.next=np
    def display(self):
        if self.head is None:
            print("empty list")
        else:
            temp=self.head
            print(temp.data,'-->',end=" ")
            while temp.next != self.head:
                temp=temp.next
                print(temp.data,'-->',end=" ")
            print(temp.next.data)
    def delete_front(self):
        if self.head is None:
            print("empty list")
        else:
            if self.head == self.tail:
                self.head=None
            else:
                temp=self.head
                self.head=temp.next
                self.tail.next=self.head
                temp.next=None
    def delete_end(self):
        if self.head is None:
            print("empty list")
        else:
            if self.head == self.tail:
                self.head=None
            else:
                temp=self.head
                temp1=self.head.next
                while temp.next != self.tail:
                    temp=temp.next
                    temp1=temp1.next
                temp.next=self.head
                self.tail=temp
                temp1=None
    def delete_position(self,pos):
        if pos==1:
            self.delete_front()
        else:
            temp=self.head
            temp1=self.head.next
            for i in range(1,pos-1):
                temp=temp.next
                temp1=temp1.next
            temp.next=temp1.next
            if temp1==self.tail:
                temp.next=self.head
                self.tail=temp
            temp1=None
        
c=CLL()
n1=Node(10)
c.head=n1
c.tail=n1
c.tail.next=c.head

n2=Node(20)
c.tail.next=n2
c.tail=n2
c.tail.next=c.head

c.insert_front(30)
c.insert_front(40)
c.insert_front(50)
c.insert_end(60)
c.insert_end(70)
c.insert_end(90)
c.insert_position(2,89)
c.insert_position(1,60)
c.insert_position(7,90000)
c.delete_front()
c.delete_front()
c.delete_front()
c.delete_end()
c.delete_end()
c.delete_end()
c.delete_end()
c.delete_position(2)
c.display()
```
