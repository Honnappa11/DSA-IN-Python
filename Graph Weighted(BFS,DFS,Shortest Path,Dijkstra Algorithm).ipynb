```python
from collections import deque
import heapq
def add_node(v):
    if v in graph:
        print(v,"it is already present in graaph")
    else:
        graph[v]=[]
def add_edges(v1,v2,cost):
    if v1 not in graph:
        print(v1,"it is not present in graph")
    if v2 not in graph:
        print(v2,"it is not present in graph")
    else:
        graph[v1].append([v2,cost])
        graph[v2].append([v1,cost])
def DFS(v,graph,visited):
    if v not in graph:
        print(v,"it is not present in graph")
    else:
        if v not in visited:
            print(v,end=" ")
            visited.add(v)
            for i in graph[v]:
                DFS(i[0],graph,visited)
def BFS(v,graph,visited):
    if v not in graph:
        print(v,"it is not present in graph")
    else:
        queue=[]
        queue.append(v)
        visited1.add(v)
        while queue:
            current=queue.pop(0)
            print(current,end=" ")
            for i in graph[current]:
                if i[0] not in visited1:
                    queue.append(i[0])
                    visited1.add(i[0])
def shortest_path(node,target,graph):
    if node not in graph:
        print(node,"it is not present in graph")
    if target not in graph:
        print(target,"it is not present in graph")
    else:
        visited={}
        queue=deque()
        visited[node]=None
        queue.append(node)
        while queue:
            current=queue.popleft()
            if current==target:
                path=[]
                while current:
                    path.append(current)
                    current=visited[current]
                return path[::-1]
            for i in graph[current]:
                v=i[0]
                if v not in visited:
                    queue.append(v)
                    visited[v]=current
def Dijkstra(graph,start):
    if start not in graph:
        print(start,"it is not present in graph")
    else:
        distance={item:float('inf')for item in graph}
        distance[start]=0
        queue=[(0,start)]
        while queue:
            current_distance,current_node=heapq.heappop(queue)
            for node,weight in graph[start]:
                new_distance=current_distance+weight
                if new_distance<distance[node]:
                    distance[node]=new_distance
                    heapq.heappush(queue,(new_distance,node))
        return distance
visited1=set()
visited=set()
graph={}
add_node("A")
add_node("B")
add_node("C")
add_node("D")
add_node("E")
add_node("F")
add_node("G")
add_node("H")
add_edges("A","B",1)
add_edges("B","C",2)
add_edges("C","D",3)
add_edges("A","E",5)
add_edges("E","F",4)
add_edges("F","H",2)
add_edges("B","G",6)
add_edges("C","G",2)
for node in graph:
    print(node,'-->',graph[node])
print('---------------------------')
print('DFS')
print()
DFS("A",graph,visited)
print()
print('-----------------------')
print('BFS')
BFS("A",graph,visited1)
print()
print('---------------------')
print("Shortest Path")
shortest_path("A","H",graph)
print('--------------------')
print('Dijkstra Algorithm')
Dijkstra(graph,"A")
```
